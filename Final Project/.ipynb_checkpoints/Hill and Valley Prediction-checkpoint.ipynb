# Hill and Valley Prediction with Logistic Regression

## Objective:

To predict whether it's a hill or a valley from a 100-point record.

Each record, when plotted on a 2 dimentional graph, will either create a hill (a "bump" in the terrain) or a valley (a "dip" in the terrain).

## Input data type:

A csv file, where each row (except the header) contains a 100 point record (V#, the # represents a number, corresponding to each point, ranging from 1 to 100, the "point" is a floating-point number) and the last column containing result (Class, where 0 means valley and 1 means hill).

Column "V1" to "V100": Represents a 100-point record. [float64]<br>
Last column "Class": 0 or 1 (i.e. valley or hill resp.) [int64]

## **Define Target Variable (y) and Feature Variables (X)**

```python
y = df['Class']

# x = df[[f'V{i + 1}' for i in range(100)]]
x = df.drop('Class', axis=1)
```

```python
y.value_counts()
```

## Data Visualization

```python
num_rows, num_cols = 10, 3

fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(10, 30))

record = 0

for i in range(num_rows):
    for j in range(num_cols):
        axs[i, j].plot(x.iloc[record])
        axs[i, j].set_title('Hill' if y[record] else 'Valley')
        axs[i, j].set_xticks([])
        axs[i, j].set_yticks([])
        record += 1
    
plt.tight_layout()
plt.show()
```

```python
# the non-ML approach
result = (x.mean(axis=1) > x.median(axis=1)) == y
result.value_counts()
```

```python
result.index[result == False]
```

```python
f'Accuracy = {1176 / (1176 + 36) * 100:.2f}%'
```

```python
# not bad, remaining 36 are a subject of case study

num_rows, num_cols = 12, 3
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(10, 30))

false_indices = (i for i in result.index[result == False])

for i in range(num_rows):
    for j in range(num_cols):
        record = next(false_indices)
        mode_y = x.iloc[record].mean()
        median_y = x.iloc[record].median()
        axs[i, j].plot(x.iloc[record])
        axs[i, j].axhline(mode_y, color='green')
        axs[i, j].axhline(median_y, color='red', linestyle='--')
        axs[i, j].set_title(f'{"H" if y[record] else "V"} mean={mode_y:.1f}, median={median_y:.1f}')
        axs[i, j].set_xticks([])
        axs[i, j].set_yticks([])

plt.tight_layout()
plt.show()
```

```python
# ohh no! check the posotion of red line!
# we been have interpreting median as a value that is equal to min+(max-min)/2,
# where in actuality it is the "middle value in a set of data"
# so, the results above are totally wrong (yes, I'm aware that mean is nothing but the average)
# let's fix it by visually check if "min + (max - min) / 2" is what we want...

num_rows, num_cols = 10, 3
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(10, 30))

ri = 0

false_indices = (i for i in result.index[result == False])

for i in range(num_rows):
    for j in range(num_cols):
        record = x.iloc[ri]
        midpoint = record.min() + (record.max() - record.min()) / 2
        axs[i, j].plot(record)
        axs[i, j].axhline(midpoint, color='green')
        axs[i, j].set_title('Hill' if y[ri] else 'Valley')
        axs[i, j].set_xticks([])
        axs[i, j].set_yticks([])
        ri += 1

plt.tight_layout()
plt.show()
```

```python
# yeap, we got it...

midpoint = x.min(axis=1) + (x.max(axis=1) - x.min(axis=1)) / 2
result = (x.mean(axis=1) < midpoint) == y  # note the comparision operator
result
```

```python
result.value_counts()
```

## Data Preprocessing

### Standadizing x

Standardization of datasets is a common requirement for many machine learning estimators implemented in sci-kit learn; they behave badly if the individual features do not (more or less) look like standardized, normally distributed data.

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
x

# अच्छा... captal x implies a vector, and lower case x represents an element of the vector
```

## Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2529, stratify=y)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Modeling

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_, model.coef_
```

## Prediction

```python
y_predict = model.predict(x_test)
```

```python
(y_predict == y_test).value_counts()
```

```python
257 / (257 + 107)
```

## Model Evaluation

```python
confusion_matrix(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```
